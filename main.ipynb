```python
peliculas=peliculas[['id', 'title', 'overview', 'genres']]
```

```python
peliculas
```

```python
peliculas['tags'] = peliculas['overview']+peliculas['genres']
```

```python
peliculas
```

```python
datos_nuevos  = peliculas.drop(columns=['overview', 'genres'])
```

```python
datos_nuevos
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(datos_nuevos['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)

similarity
```

```python
datos_nuevos[datos_nuevos['title']=="La Era de Hielo: Las aventuras de Buck"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(datos_nuevos.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=datos_nuevos[datos_nuevos['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(datos_nuevos.iloc[i[0]].title)
```

```python
recommand("Sangre y oro")
```
